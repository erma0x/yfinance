### yfinance python library examples 

library: https://pypi.org/project/yfinance/

```python
import yfinance as yf
```

```python
aapl= yf.Ticker("aapl")
print(aapl)
```

```python
aapl.info
```

```python
aapl_historical = aapl.history(start="2022-03-10", end="2022-03-16", interval="1m")
aapl_historical
```

```python
data = yf.download("AMZN AAPL GOOG", start="2017-01-01", end="2017-04-30")
data
```

```python
data = yf.download("AMZN AAPL GOOG", start="2017-01-01",
                    end="2017-04-30", group_by='tickers')
data
```
